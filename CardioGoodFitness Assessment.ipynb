### Multivariate Analysis

```python
#Comparing our income distribution with the Age
sns.catplot(data=df, x = 'Age', y = 'Income', kind = 'box')
plt.title('Income Distribution Accourding to Age')
plt.ylabel('Income')
plt.xticks(rotation = 90)
plt.show()
```

We can see from this chart that people who are aged 30 and 47 years have the highest income stream with people below the age of 21 had the lowest income stream.

```python
#Comparing our Product distribution with the Age
sns.catplot(data=df, x = 'Age', y = 'Product', kind = 'box')
plt.title('Product Distribution Accourding to Age')
plt.ylabel('Product')
plt.show()
```

From this distribution we can see that TM195 has the highest demand from the retail store, while TM798 has least demand from the retail store.

```python
#Comparing our Marital Status distribution with the Age
sns.boxplot(data=df, x = 'Product',y = 'Age', hue = 'MaritalStatus')
plt.title('Product Distribution According to Age')
plt.ylabel('Product')
plt.show()
```

From our distribution we can see that the highest consumers of the product are mostly people who have partners. 

```python
#Comparing our Fitness Aim distribution with the Age
sns.boxplot(data=df, x = 'Fitness',y = 'Age')
plt.title('Fitness Aim According to Age')
plt.ylabel('Age')
plt.show()
```

From our chart we can see that the age range with the higher cluster of age groups with fitness goals are between 23 and 33 years of age. Also, people who chose 3 and 5 as their fitness goals spread across all ages.

```python
#Comparing the Usage against the Age
sns.boxplot(data=df, x = 'Usage',y = 'Age')
plt.title('Usage Against Age')
plt.ylabel('Age')
plt.show()
```

From our chart we can see that people between the ages of 23 and 35 said they will use the training equipment 3 times, which is the highest, while people who said they will use the equipment 7 times are had the least.

```python
sns.boxplot(data=df, x = 'Fitness',y = 'Age',hue = 'Product')
plt.title('Fitness According to Age and Product Preferences')
plt.ylabel('Age')
plt.show()
```

From our chart we can that TM195 has spread among all fitness goal segments, followed by TM498  and lastly by TM798 and the largest age group that makes use of the products are clustered between 23 and 35. 

```python
sns.scatterplot(data=df, x = 'Income',y = 'Age',hue = 'Product')
plt.title('Income According to Age and Product Preferences')
plt.ylabel('Age')
plt.show()
```

From our chart we can see the people whose income is 47k and 100k+ patronize TM798, while people whose income is between 32k and 67k patronize more of TM498, while people whose income is between 29k and 67k also patronize TM195.
